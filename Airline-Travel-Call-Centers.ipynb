```python
import spacy

# Load SpaCy's pre-trained NER model
nlp = spacy.load("en_core_web_sm")

# Sample text data
text = "The customer called yesterday to complain about the long wait times."

# Process the text with SpaCy NER model
doc = nlp(text)

# Extract entities for each word in the sentence
word_entities = [(token.text, token.ent_type_) for token in doc]

# Print word-level entities
for word, entity in word_entities:
    print(f"Word: {word}, Entity: {entity if entity else 'O'}")
```

```python
from transformers import pipeline

# Load NER pipeline with BERT model
ner_pipeline = pipeline("ner", model="bert-base-uncased", tokenizer="bert-base-uncased")

# Sample text data
text = "The customer called yesterday to complain about the long wait times."

# Perform NER on the text
ner_results = ner_pipeline(text)

# Print named entities
for result in ner_results:
    print(f"Entity: {result['entity']}, Score: {result['score']}, Indexes: {result['start']}-{result['end']}")
```

```python
import spacy

# Load SpaCy's English language model
nlp = spacy.load("en_core_web_sm")

# Sample conversation text
conversation_text = """
Certainly! Here's a new paragraph tailored to an airline customer care scenario:

"Good morning! This is Emily from SkyLink Airlines. Am I speaking with Mr. Johnson? Great, Mr. Johnson. How can I assist you today? Ah, I see you have a concern about your upcoming flight reservation. Could you please provide me with your booking reference number? Thank you. I've pulled up your reservation, and I understand that you're experiencing some issues with your seating arrangement. Is that correct? I apologize for any inconvenience caused, Mr. Johnson. Let me check the available options for you. It seems like there's been a last-minute change in the aircraft configuration, resulting in a shift in your seat assignment. However, rest assured, I'll do my best to find a suitable alternative for you. Please bear with me for a moment while I explore our options. Thank you for your patience, Mr. Johnson. I've managed to secure a seat for you in the preferred zone with extra legroom. How does that sound? Excellent! I've gone ahead and updated your reservation accordingly. Is there anything else I can assist you with today, Mr. Johnson? If not, I wish you a pleasant journey with SkyLink Airlines. Thank you for choosing us, and have a wonderful day ahead!"
"""

# Process the conversation text using SpaCy NER
doc = nlp(conversation_text)

# Initialize variables to store feature magnitudes
time_duration_magnitude = 0
customer_satisfaction_mention_magnitude = 0
transfer_action_magnitude = 0

# Iterate over the entities extracted by SpaCy NER
for ent in doc.ents:
    if ent.label_ == "TIME":
        # If entity is a time duration
        time_duration_magnitude += 1
    elif ent.text.lower() in ["satisfied", "unsatisfied", "happy", "unhappy", "pleased", "dissatisfied"]:
        # If entity is a customer satisfaction mention
        customer_satisfaction_mention_magnitude += 1
    elif ent.text.lower() in ["transfer", "transferred", "transferring"]:
        # If entity is a transfer action
        transfer_action_magnitude += 1

# Print the magnitude of each feature
print("Magnitude of Time Duration Feature:", time_duration_magnitude)
print("Magnitude of Customer Satisfaction Mention Feature:", customer_satisfaction_mention_magnitude)
print("Magnitude of Transfer Action Feature:", transfer_action_magnitude)
```

```python
from transformers import pipeline

def calculate_fcr_magnitude(conversation):
    # Initialize sentiment analysis pipeline
    sentiment_pipeline = pipeline("sentiment-analysis")

    # Analyze sentiment of the conversation
    sentiment_result = sentiment_pipeline(conversation)[0]

    # Determine FCR magnitude based on sentiment
    if sentiment_result['label'] == 'POSITIVE':
        fcr_magnitude = 100  # Assume positive sentiment indicates issue resolution
    else:
        fcr_magnitude = 0    # Assume negative sentiment indicates unresolved issue

    return fcr_magnitude

# Example conversation text
conversation_text = """
Thank you for calling. I'm glad to inform you that we've successfully resolved your issue during this call.
"""

# Calculate the FCR magnitude for the given conversation
fcr_magnitude = calculate_fcr_magnitude(conversation_text)
print("FCR Magnitude:", fcr_magnitude)
```

<h2>Pie chart for Call Resolution Rate:</h2>


```python
import matplotlib.pyplot as plt

# Call-Level KPIs
call_resolution_rate = 75  # Example value

# Pie chart for Call Resolution Rate
plt.figure(figsize=(6, 6))
plt.pie([call_resolution_rate, 100 - call_resolution_rate], labels=["Resolved", "Not Resolved"], autopct='%1.1f%%')
plt.title("Call Resolution Rate")
plt.show()
```
